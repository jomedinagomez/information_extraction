```python
from unstructured.partition.pdf import partition_pdf
from unstructured.chunking.title import chunk_by_title
import htmltabletomd
import pandas as pd

# Returns a List[Element] present in the pages of the parsed pdf document
elements = partition_pdf("MICROSOFT-10Q-FY2022-Q3.pdf")
```

```python
chunks = chunk_by_title(elements, multipage_sections=True, new_after_n_chars=1500, combine_text_under_n_chars=500)
for chunk in chunks:
    print("-----------------------------------------------------------------------------------")
    print("-----------------------------------------------------------------------------------")
    print(chunk.metadata.to_dict())
    print("----------------------------------------------------------------")
    if chunk.category == "Table":
        #print(chunk.metadata.text_as_html)
        try:
            print(pd.read_html(chunk.metadata.text_as_html)[0])
        except:
            print(chunk.metadata.text_as_html)
    else:
        print(chunk.text)
```
